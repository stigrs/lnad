```python
import networkx as nx
import igraph as ig
import matplotlib.pyplot as plt
import lnad.analysis as lnad
import lnad.dismantling as lnad_dismantling

G = nx.petersen_graph()
N = lnad.NetworkAnalysis(ig.Graph.from_networkx(G))

_, ax = plt.subplots()
N.plot(target=ax)

print(N.degree_centrality())
print(N.betweenness_centrality())
print(N.eigenvector_centrality())
print(N.closeness_centrality())
print(N.articulation_points())
print(N.global_efficiency())
print(N.largest_connected_component())
```

```python
d = N.degree_centrality()
print(list(d.values())[0])
```

```python
D = lnad_dismantling.NetworkDismantling(N.graph)

G_attacked, n_attacked, lcc, slcc, eff, c = D.node_iterative_centrality_attack(
    nattacks=3, centrality_method=lnad.degree_centrality
)
print(n_attacked)
print(lcc)
```

```python
_, ax = plt.subplots()
ig.plot(G_attacked, target=ax)
```
